# Graphs partial global warming potential over time

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import make_interp_spline, BSpline
```

## Plot partial GWP100(t)

```python
df = pd.read_excel('partialGWP.xlsx', 'GWP100_CCincl.')

fig, ax = plt.subplots(figsize=(7, 5))

x = df['Year']
xnew = np.linspace(df['Year'].min(), df['Year'].max(), 1000) 

# smoothen time differentiated GPW curves
spl_bioref = make_interp_spline(df['Year'], df['Biorefinery- time differentiated'], k=7)  # type: BSpline
y_bioref_smooth = spl_bioref(xnew)

spl_comb = make_interp_spline(df['Year'], df['CHP - time differenciated'], k=7)  # type: BSpline
y_comb_smooth = spl_comb(xnew)

plt.plot(xnew, y_comb_smooth, label='Combustion', color = '#4363d8')
plt.plot(xnew, y_bioref_smooth, label='Biorefinery', color = '#e6194B')

plt.scatter(x, df['Biorefinery - aggregated'], marker = 'o', s = 2, color = '#e6194B')
plt.scatter(x, df['CHP - aggregated'], marker = 'o', s = 2, color = '#4363d8')

# Adding titles and labels
plt.xlabel('Year')
plt.ylabel('Partial GWP100 (kg CO2-eq/FU)')
plt.legend()

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.tight_layout()
plt.savefig('partialGWP100.pdf')

# Show plot
plt.show()
```
